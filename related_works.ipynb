# Related Works and Findings

## Direct Dataset Findings
- Generating and Personalizing Bundle Recommendations on Steam. https://cseweb.ucsd.edu/~jmcauley/pdfs/sigir17.pdf. Predicted bundles through using Bayesian Ranking and latent factors.
- Item Recommendation on Monotonic Behavior Chains. https://cseweb.ucsd.edu/~jmcauley/pdfs/recsys18b.pdf. Predicting monotonic action sets using Bayesian Ranking and latent factors. One dataset of many.
- Self-Attentive Sequential Recommendation. https://cseweb.ucsd.edu/~jmcauley/pdfs/icdm18.pdf. Guessing sequential/monotonic actions using latent factors again but with self-attention.
- Game Recommendation Analysis Using Steam Profiles and Reviews. https://scholar.smu.edu/cgi/viewcontent.cgi?article=1272&context=datasciencereview. Data analysis paper that found out relationships in genre, popularity in games and recommendation.
- https://github.com/TianyaoHan/Steam-Recommendation-System. A github project for game recommendation that uses that dataset alongside collaborative filtering. Unfortunately does not include findings or evaluation.
- Building Recommender Systems for Video Games on Steam. https://library.ucsd.edu/dc/object/bb5021836n/_3_1.pdf. A hybrid appraoch of feature based and collaborative filtering.

## Indirect Dataset Findings
- Hybrid system for video game recommendation based on implicit ratings and social networks. https://2024.sci-hub.se/7994/446779116f080fce5fb45dd5e6c769e1/10.1007@s12652-020-01681-0.pdf. A hybrid approach of features and collaborative filtering for games
- Machine / Deep Learning Model for a recommender engine. https://www.linkedin.com/pulse/recommender-engine-steam-dataset-david-adamson-mbcs/. A pretty standard approach but it factors in time played into the prediction.
- https://github.com/thehir0/steam-recsys?tab=readme-ov-file. https://github.com/AdrianKlessa/SteamScout. Github projects that use LightFM to create hybrid/ensemble approaches.
